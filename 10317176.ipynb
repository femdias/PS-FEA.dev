Notamos que alguns valores das colunas não condizem com a realidade, como o curso Ciência das Lulas. Antes de arrumar isso, devemos tirar os 'buracos' (NaN) da base, substituindo eles por 'Missing'. Depois, podemos analisar os valores estranhos e substituí-los também por 'Missing'.

```python
# Para eliminar os dados faltantes, podemos usar o método .fillna
base_comp = base.fillna('Missing')

#Podemos ver que todos os 'buracos' foram substituidos com 'Missing', todas colunas tem 308 de contagem.
base_comp.describe() 
```

```python
#Construindo gráficos para todas as variáveis por meio de itereção
for i in (base_comp.columns):
    plt.figure(figsize=(16, 6)) #Editando o tamanho dos gráficos
    print(sns.countplot(base[i]))
```

Podemos notar alguns erros: "Ciência das Lulas em 'curso', dois valores  iguais para o 'periodo' (diu e Diurno, not e Noturno), valores estranhos (-1337, -40, 10000, 999, -10000) e com 'km/s^2'  e  na coluna 'ano'. 

Precisamos fazer outro tipo de análise para avaliar as últimas 5 colunas.

```python
#tirando esse estranho 'km/s^2' que aparece na coluna de ano
base_comp['ano'] = base_comp['ano'].str.rstrip(' km/s^2')
base_comp['ano'].head()
```

```python
#Fazendo uma iteração para arrumar as colunas

for i in np.arange(len(base_comp)):
    if base_comp.loc[base_comp.index[i],'ano']  not in ['1º','2º','3º','4º','5º','6º','7º','8º','9º','10º']:
        base_comp.loc[base_comp.index[i], 'ano'] = "Missing"
        
    if base_comp.loc[base_comp.index[i],'curso']  == 'Ciência das Lulas':
        base_comp.loc[base_comp.index[i], 'curso'] = "Missing"
        
    if base_comp.loc[base_comp.index[i],'periodo']  == 'diu':
        base_comp.loc[base_comp.index[i], 'periodo'] = "Diurno"
    elif base_comp.loc[base_comp.index[i],'periodo']  == 'not':
        base_comp.loc[base_comp.index[i], 'periodo'] = "Noturno"
        
#Mostrando apenas os valores únicos das colunas para ver se agora não tem nenhum valor estranho
print(set(base_comp['ano']))
print(set(base_comp['curso']))
print(set(base_comp['periodo']))
```

```python
#Vamos agora arrumar a coluna 'motivo_interesse'
# Para isso, criaremos várias variáveis dummy (1 ou 0, Sim ou Não)

#primeiro, precisamos achar quais são as opções de interesse
lista = []
for i in np.arange(len(base_comp)):
    listinha = base_comp.loc[base_comp.index[i],'motivo_interesse'].split(', ') #Separando as opcões
    lista = np.append(lista, listinha)

#Usando set para achar os valores únicos
lista_unica = set(lista)
lista_unica
```

```python
#Agora, podemos criar as dummies. Para isso, iteramos os interesses (j) e as linhas (i) para marcar os interesses que aparecem 
# em cada indivíduo, através de 1 (Sim) e 0 (Não) nas dummies.

#Criando um DataFrame de 0's com as colunas nomeadas com as opções de motivo de interesse
interesses = pd.DataFrame(columns = lista_unica, index = np.arange(len(base_comp))).fillna(0)

#Se o motivo está contido na coluna 'motivo_interesse', será marcado com 1 em sua coluna
for j in lista_unica:
    for i in np.arange(len(base_comp)):
        if  j in base_comp.loc[base_comp.index[i],'motivo_interesse']:
            interesses.loc[interesses.index[i],j] = 1
        else:
            interesses.loc[interesses.index[i],j] = 0
            
#Para não haver varias colunas com o mesmo nome (nas bases que virão a seguir),vamos renomear 'Missing'
interesses = interesses.rename(columns = {'Missing': 'Missing_Interesse'})

interesses.head(10)
```

```python
#Com esse DataFrame 'interesses' podemos ver o interesse de cada indivíduo com mais facilidade
#Podemos plotar um gráfico simples que mostra a quantidade de pessoas que tem interesse em cada opção

inter = interesses.sum()
plt.figure(figsize=(16, 6))
inter.plot(kind='bar')
```

```python
#Vamos juntar algumas colunas que são a mesma coisa

sql = linguagem['SQL'] + linguagem['Sql '] + linguagem['Sql'] + linguagem['sql'] + linguagem['MySQL']
for i in np.arange(len(sql)):
    if sql[i] >= 2:
        sql[i] = 1
linguagem['SQL'] = sql

html = linguagem['Html'] + linguagem['HTML']
for i in np.arange(len(html)):
    if html[i] >= 2:
        html[i] = 1
linguagem['HTML'] = html

Go = linguagem['golang'] + linguagem['Go'] + linguagem['Golang'] + linguagem['golang']
for i in np.arange(len(html)):
    if Go[i] >= 2:
        Go[i] = 1
linguagem['Go'] = Go

Js = linguagem['JavaScript'] + linguagem['Javascript']
for i in np.arange(len(html)):
    if Js[i] >= 2:
        Js[i] = 1
linguagem['JavaScript'] = Js

#Go.value_counts()

linguagem = linguagem.drop(['Sql ', 'Sql', 'sql', 'MySQL', 'Html', 'Golang', 'golang', 'Javascript'] , axis=1 )
linguagem.columns
```

```python
# Arrumando a coluna 'forma_aprender'

lista_forma = []
for i in np.arange(len(base_comp)):
    listinha_forma = base_comp.loc[base_comp.index[i],'forma_aprender'].split(', ')
    lista_forma = np.append(lista_forma, listinha_forma)
    
lista_forma_unica = set(lista_forma)
print(lista_forma_unica)
```

```python
forma = pd.DataFrame(columns = lista_forma_unica, index = np.arange(len(base_comp))).fillna(0)
for j in lista_forma_unica:
    for i in np.arange(len(base_comp)):
        if  j in base_comp.loc[base_comp.index[i],'forma_aprender']:
            forma.loc[linguagem.index[i],j] = 1
        else:
            forma.loc[linguagem.index[i],j] = 0
            
#Renomeando a coluna 'Missing'
forma = forma.rename(columns = {'Missing': 'Missing_forma', 'Não Sei': 'Não Sei Forma'})

forma.head()
```

```python
#Vamos juntar a coluna 'Fazendo', 'Praticando' e 'Fazendo/Praticando'
fazendo = forma['Fazendo'] + forma['Praticando'] + forma['Fazendo/Praticando']
for i in np.arange(len(fazendo)):
    if fazendo[i] >= 2:
        fazendo[i] = 1
forma['Fazendo'] = fazendo

presen = forma['Aula presencial'] + forma['Presencial: Prática + teoria ']
for i in np.arange(len(presen)):
    if presen[i] >= 2:
        presen[i] = 1
forma['Aula Presencial'] = presen

forma = forma.drop(['Fazendo/Praticando' , 'Praticando', 'Aula presencial', 'Presencial: Prática + teoria '] , axis=1 )
forma.columns
```

```python
#Agora que as colunas finalmente estão separadas, podemos juntar todos os DataFrames

base_completona = pd.concat([base_comp, interesses, area, tema, linguagem, forma], axis=1)
pd.set_option('display.max_columns', None)
base_completona
```

```python
#Agora, iremos compararam os Motivos de Interesse por programação divididos pelos cursos.

#Selecionando só os valores mais importantes de 'curso' e as colunas mais com mais ocorrencias dos motivos de interesse
#Fazemos isso para deixar os gráficos menos poluídos
curso = ['Atuárias', 'Contabilidade', 'Economia', 'Administração']
inte = ['Mercado de Trabalho' , 'Curiosidade' , 'Pesquisa Acadêmica', 'Desenvolvimento Web']

#Construindo um gráfico para cada curso 
for j in curso:
    base_filtrada =  base_completona[base_completona['curso'] == j]
    df = pd.DataFrame()
    for i in inte:
        xx = base_filtrada[i]
        df[i] = xx
    plt.figure(figsize=(16, 6))
    sns.barplot(x = (df.sum()).index, y = df.sum()/ df.count()).set_title(j)
```
